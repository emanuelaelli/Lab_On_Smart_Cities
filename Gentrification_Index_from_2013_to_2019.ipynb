# Componenti indice di gentrificazione

L'equazione che costituisce la base della nostra analisi ed è stata generata per calcolare il punteggio di gentrificazione di una Super Output Area (LSOA) di livello inferiore è la seguente: 

image.png


Dove:
* **G** rappresenta l'indice di gentrificazione, costituendo una scala da 0,0 a 1,0 dove il limite inferiore indica un'area di minore gentrificazione e il limite superiore indica un'area di grave gentrificazione tra il 2010 e il 2016
c rappresenta il tasso di abbandono della popolazione a livello familiare all'interno di un'area tra l'inizio del 2010 e la fine del 2016, o "un rapporto tra le famiglie che sono cambiate in ogni LSOA... tra [2010 e] 2016" (CDRC, 2016) .

* **c** rappresenta il tasso di abbandono della popolazione a livello familiare all'interno di un'area tra l'inizio del 2010 e la fine del 2016, o "un rapporto tra le famiglie che sono cambiate in ogni LSOA... tra [2010 e] 2016" (CDRC, 2016) 

* **e** rappresenta la variazione relativa della proporzione di residenti non bianchi a livello individuale all'interno di un'area

* **h** rappresenta la variazione relativa del prezzo medio di vendita della casa rispetto alla media del distretto tra dicembre 2009 e dicembre 2016 (ovvero il prezzo medio di vendita della casa per un dato LSOA era il 90% della media del distretto nel 2009 ed è aumentato a 110 per centesimo della media a livello di distretto nel 2016, producendo un punteggio di 0,22 utilizzando la formula della variazione relativa)
  
* **d** rappresenta la variazione relativa del punteggio dell'indice di deprivazione multipla (IMD) tra il 2010 e il 2015
  
* viene aggiunto un **b-value** di 0,25 per standardizzare il punteggio di gentrificazione tra 0,0 e 1,0
